```python
from geopy.geocoders import Nominatim 
import requests
import folium
import xml
from pandas.io.json import json_normalize
import matplotlib.cm as cm
import matplotlib.colors as colors
from sklearn.cluster import KMeans
from bs4 import BeautifulSoup

!conda install -c conda-forge folium=0.5.0 --yes
```

```python
url = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
soup = BeautifulSoup(url,'lxml')
```

```python
table_post = soup.find('table')
fields = table_post.find_all('td')
postc = []
borough = []
neighbourhd = []
for i in range(0, len(fields), 3):
    postc.append(fields[i].text.strip())
    borough.append(fields[i+1].text.strip())
    neighbourhd.append(fields[i+2].text.strip())
        
df_pc = pd.DataFrame(data=[postc, borough, neighbourhd]).transpose()
df_pc.columns = ['Postcode', 'Borough', 'Neighbourhood']
df_pc.head()
```

```python
df_pc['Borough'].replace('Not assigned', np.nan, inplace=True)
df_pc.dropna(subset=['Borough'], inplace=True)
df_pc.head()
```

```python
df_pcn = df_pc.groupby(['Postcode', 'Borough'])['Neighbourhood'].apply(', '.join).reset_index()
df_pcn.columns = ['Postcode', 'Borough', 'Neighbourhood']
df_pcn['Neighbourhood'].replace('Not assigned', "Queen's Park", inplace=True)
df_pcn
```

```python
df_pcn.shape
```

```python
df_geo = pd.read_csv('http://cocl.us/Geospatial_data')

df_geo.columns = ['Postcode','Latitude','Longitude']
df_postcode = pd.merge(df_pcn,df_geo, on=['Postcode'],how='inner')
df_toronto = df_postcode[['Borough', 'Neighbourhood','Postcode','Latitude','Longitude']].copy()
df_toronto.head()
```

```python
address = 'Toronto,Canada'
geolocator = Nominatim()
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The Geograpical coordinates of Toronto City are {}, {}.'.format(latitude, longitude))
```

```python
# create map of Toronto
map_toronto = folium.Map(location=[latitude, longitude],zoom_start=10)

# add markers to map
for lat, lng, borough, neighborhood in zip(df_toronto['Latitude'],df_toronto['Longitude'],df_toronto['Borough'],df_toronto['Neighbourhood']):
    label = '{}, {}'.format(neighborhood,borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
    [lat,lng],
    radius=4,
    popup=label,
    color='red',
    fill=True,
    fill_color='#3188cc',
    fill_opacity=0.5,
    parse_html=False).add_to(map_toronto)  
    
map_toronto
```
